# Linear Regression

## Step by Step evaluation of precision

```python
import numpy as np
import hnumpy as hnp
import matplotlib.pyplot as plt
```

### np.mean

```python
N = 32

def mean(x):
    return np.mean(x)

he_mean = hnp.compile_fhe(
    mean,
    { "x": hnp.encrypted_ndarray(bounds=(0,1), shape=(N,)) }
)

expected = []
simulate = []
rounds = 1000
for r in range(rounds):
    x = np.random.uniform(0,1,N)
    
    exp_res = mean(x)
    sim_res = he_mean.simulate(x)
    
    expected.append(exp_res)
    simulate.append(sim_res)
    
plt.plot(expected,simulate, "o", markersize=2)
_x = np.linspace(min(expected),max(expected),100)
plt.plot(_x,_x)
plt.grid()
plt.show()
```

### y - np.mean(y)

```python
N = 32

def getY(X):
    return 0.2 + X * 0.42 + np.random.uniform(-0.025, 0.025, len(X))

expected = []
simulate = []

rounds = 100
for r in range(rounds):
    X = np.linspace(0,1,N)
    y = getY(X)
    
    def k(y):
        return y-np.mean(y)
    
    he_k = hnp.compile_fhe(
        k,
        { "y": hnp.encrypted_ndarray(bounds=(0,1), shape=(N,)) }
    )
    
    exp_res = np.sum(np.abs(k(y)))
    sim_res = np.sum(np.abs(he_k.simulate(y)))
    
    expected.append(exp_res)
    simulate.append(sim_res)
    
plt.plot(expected,simulate, "o", markersize=3)
_x = np.linspace(min(expected),max(expected),100)
plt.plot(_x,_x)
plt.grid()
plt.show()
```

### (x - mean(x)) * (y - mean(y))

```python
N = 32

def getY(X):
    return 0.2 + X * 0.42 + np.random.uniform(-0.025, 0.025, len(X))

expected = []
simulate = []

rounds = 100
for r in range(rounds):
    X = np.linspace(0,1,N)
    y = getY(X)
    
    def k(y):
        return np.multiply(X-np.mean(X),y-np.mean(y))
    
    he_k = hnp.compile_fhe(
        k,
        { "y": hnp.encrypted_ndarray(bounds=(0,1), shape=(N,)) }
    )
    
    exp_res = np.sum(np.abs(k(y)))
    sim_res = np.sum(np.abs(he_k.simulate(y)))
    
    expected.append(exp_res)
    simulate.append(sim_res)
    
plt.plot(expected,simulate, "o", markersize=2)
_x = np.linspace(min(expected),max(expected),100)
plt.plot(_x,_x)
plt.grid()
plt.show()
```

```python
N = 32

def getY(X):
    return 0.2 + X * 0.42 + np.random.uniform(-0.025, 0.025, len(X))

expected = []
simulate = []

rounds = 100
for r in range(rounds):
    X = np.linspace(0,1,N)
    y = getY(X)
    
    def k(y):
        return np.sum(np.multiply(X-np.mean(X),y-np.mean(y)))
    
    he_k = hnp.compile_fhe(
        k,
        { "y": hnp.encrypted_ndarray(bounds=(0,1), shape=(N,)) }
    )
    
    exp_res = k(y)
    sim_res = he_k.simulate(y)
    
    expected.append(exp_res)
    simulate.append(sim_res)
    
plt.plot(expected,simulate, "o", markersize=3)
_x = np.linspace(min(expected),max(expected),100)
plt.plot(_x,_x)
plt.grid()
plt.show()
```
