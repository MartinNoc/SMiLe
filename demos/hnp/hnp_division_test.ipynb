# HNP Division Precision Benchmark

### Division by a constant

```python
import hnumpy as hnp
import numpy as np
import matplotlib.pyplot as plt
```

```python
inp_len = 100000

d_max = 5
n_max = 12

fig, (ax1, ax2) = plt.subplots(2, sharex=True, figsize=(10,8))
x_axis = [2**n for n in range(1,n_max+1)]

fig.suptitle('Simulated division error')

for d in range(1,d_max + 1):
    D = 2**d
    
    def div_const(x):
        return np.divide(x,D)

    abs_errors = []
    rel_errors = []
    for n in range(1,n_max + 1):
        N = 2**n

        he_div_const = hnp.compile_fhe(
            div_const,
            {
                "x": hnp.encrypted_ndarray(bounds=(0,N), shape=(inp_len,)),
            }
        )

        x = np.random.uniform(0, N, (inp_len,))
        s_result = he_div_const.simulate(x)
        abs_err = np.abs(s_result - x/D)
        rel_err = 100 * np.divide(abs_err, (x/D))
        
        abs_errors.append(np.mean(abs_err))
        rel_errors.append(np.mean(rel_err))

    ax1.plot(x_axis, abs_errors, "o-", label=D)
    ax2.plot(x_axis, rel_errors, "o--", label=D)

ax1.set(xlabel='input upper bound', ylabel='absolute error')
ax1.label_outer()
ax1.grid()
ax1.legend(title="divisor")

ax2.set(xlabel='input upper bound', ylabel='relative error (%)')
ax2.label_outer()
ax2.grid()
ax2.legend(title="divisor")

fig.show()
    
# plt.plot(x_axis, abs_errors, "o-")
# plt.ylabel("absolute error")
# plt.xlabel("upper input bound")
# plt.grid()
# plt.show()

# plt.plot(x_axis, rel_errors, "o-")
# plt.ylabel("relative error [%]")
# plt.xlabel("upper input bound")
# plt.grid()
# plt.show()
```

### 1/x

```python
def inverse(x):
    return np.reciprocal(x)
```

```python
inp_len = 1000000

n_max = 8

fig, (ax1, ax2) = plt.subplots(2, sharex=True, figsize=(10,8))
x_axis = [2**n for n in range(1,n_max+1)]

fig.suptitle('Simulated inverse error')

abs_errors = []
rel_errors = []
for n in range(1,n_max + 1):
    N = 2**n

    he_inv = hnp.compile_fhe(
        inverse,
        {
            "x": hnp.encrypted_ndarray(bounds=(1,N), shape=(inp_len,)),
        }
    )

    x = np.random.uniform(1, N, (inp_len,))
    s_result = he_inv.simulate(x)
    abs_err = np.abs(s_result - inverse(x))
    rel_err = 100 * np.divide(abs_err, inverse(x))

    abs_errors.append(np.mean(abs_err))
    rel_errors.append(np.mean(rel_err))

ax1.plot(x_axis, abs_errors, "o-", label=D)
ax2.plot(x_axis, rel_errors, "o--", label=D)

ax1.set(xlabel='input upper bound', ylabel='absolute error')
ax1.label_outer()
ax1.grid()
#ax1.legend(title="divisor")

ax2.set(xlabel='input upper bound', ylabel='relative error (%)')
ax2.label_outer()
ax2.grid()
#ax2.legend(title="divisor")

fig.show()
```
